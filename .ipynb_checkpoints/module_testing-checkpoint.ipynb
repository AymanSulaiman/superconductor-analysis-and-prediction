```python
import pandas as pd
import os
from tensorflow_model_and_validation import dl_model
```

```python
df = pd.read_csv(os.path.join('data','train.csv'))
df
```

```python
X = df.drop(columns=['critical_temp'])
X
```

```python
y = df[['critical_temp']].values.reshape(-1,1)
y
```

```python
deep_learning = dl_model(X=X,y=y)
```

```python
X_train, X_test, y_train, y_test = deep_learning.train_test_split()
```

```python
model = deep_learning.model_itself(X_train, y_train)
model
```

```python
analysis = dl_model.model_analysis(model=model, X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test)
analysis
```
